```python
import numpy as np
import tensorflow as tf
import tf_keras as keras
from tf_keras.models import Sequential
from tf_keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
import tensorflowjs as tfjs
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
!pip show keras
```

```python
import os
os.environ['TF_USE_LEGACY_KERAS'] = '1'
```

```python
def load_json_to_numpy(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)

    return np.array([[item["x"], item["y"], item["z"]] for item in data], dtype=np.float64)

still_data = load_json_to_numpy('data/still_data.json')
walking_data = load_json_to_numpy('data/walking_data.json')
attacked_data = load_json_to_numpy('data/attacked_data.json')
jogging_data = load_json_to_numpy('data/jogging_data.json')
```

```python
def plot_accelerometer_data(data, title="Accelerometer Data"):
    """
    Plot the accelerometer data over time.
    """
    time = np.arange(len(data)) / 100.0  # Convert to seconds (100Hz sampling rate)

    plt.figure(figsize=(10, 4))
    plt.plot(time, data[:, 0], label='X-axis', alpha=0.8)
    plt.plot(time, data[:, 1], label='Y-axis', alpha=0.8)
    plt.plot(time, data[:, 2], label='Z-axis', alpha=0.8)

    plt.axhline(1, color='gray', linestyle='--', linewidth=1, label='Gravity (1g on Z)')

    plt.xlabel("Time (seconds)")
    plt.ylabel("Acceleration (g)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

# Example Usage:
time_steps = 500  # 10 seconds of data at 100Hz
features = 3       # X, Y, Z axes

normal_data = np.concatenate((still_data, walking_data, jogging_data))
fight_data = attacked_data

plot_accelerometer_data(normal_data, title="Normal Activity Data")
plot_accelerometer_data(fight_data, title="Fight Activity Data")
```

```python
# --- Parameters ---
TIME_STEPS = 500    # Number of time steps in each sample
FEATURES = 3       # Accelerometer axes: x, y, z
FIGHT_SAMPLES = int(0.1*fight_data.shape[0])
NORMAL_SAMPLES = int(0.1*normal_data.shape[0])
print(FIGHT_SAMPLES)
print(NORMAL_SAMPLES)

# --- Generate Dataset ---
data = []
labels = []

# Normal activity samples (label 0)
for _ in range(NORMAL_SAMPLES):
    random_start = np.random.randint(0, len(still_data) - TIME_STEPS)
    data.append(normal_data[random_start: random_start + TIME_STEPS].tolist())
    labels.append(0)

# Fight activity samples (label 1)
for _ in range(FIGHT_SAMPLES):
    random_start = np.random.randint(0, len(fight_data) - TIME_STEPS)
    data.append(fight_data[random_start: random_start + TIME_STEPS].tolist())
    labels.append(1)

data = np.array(data)    # Shape: (2000, 50, 3)
labels = np.array(labels)
```

```python


# Shuffle and split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data, labels, test_size=0.2, random_state=42, shuffle=True
)

# --- Define the Model ---
model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(TIME_STEPS, FEATURES)),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# --- Train the Model ---
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# --- Evaluate the Model ---
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.2f}")

# --- Save the Model ---
# This can later be converted to a mobile-friendly format (e.g., TensorFlow Lite) for deployment.
model.save('model/model.h5')
tfjs.converters.save_keras_model(model, 'model/')
```

```python
# use model on test data
one_sample = X_test[1]
one_sample = np.expand_dims(one_sample, axis=0)
print(one_sample.shape)
one_sample_label = y_test[1]
print(one_sample_label)
print(model.predict(one_sample))
```
